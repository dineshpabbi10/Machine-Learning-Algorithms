```python
import matplotlib.pyplot as pp
import numpy as np
```

```python
x = np.array([1,7,3,11,20,9])
pp.plot(x,x**2,color="red",marker = "^")
pp.xlabel("Numbers")
pp.ylabel("Square")
pp.title("Scatter plot of numbers vs their square")
pp.show()
```

```python
y = np.random.random([100,2])
for x in range(np.shape(y)[0]):
    pp.scatter(y[x][0],y[x][1],color= "orange",marker="^")
    pp.title("Random Scatter Plot")
    pp.xlabel("x axis")
    pp.ylabel("y axis")
pp.show()
    
```
