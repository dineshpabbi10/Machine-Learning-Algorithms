```python
import matplotlib.pyplot as pp
import numpy as np
```

# Without relation between x and y

```python
mean = np.array([0.00,0.00])
cov = np.array([[1,0],[0,1]])
n = np.random.multivariate_normal(mean,cov,500)
# print(n)
pp.scatter(n[:,0],n[:,1])
pp.show()
```

# With relation between x and y in covariance matrix:

```python
mean2 = np.array([0,0])
cov2 = np.array([[1,0.5],[0.5,1]])
n2 = np.random.multivariate_normal(mean2,cov2,500)
# print(n)
pp.scatter(n2[:,0],n2[:,1])
pp.show()
```
