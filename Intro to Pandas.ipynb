## Calculating length of movies names and frequency

```python
title = data.get("movie_title")
    #Get the data in title variable
print(type(title))
title =  list(title)
k =  dict()
for x in title:
    try:
        k[len(x)] += 1
    except:
        k[len(x)] = 1
print(k)
```

# Plotting a scatter plot

```python
pp.scatter(k.keys(),k.values())
pp.show()
```
