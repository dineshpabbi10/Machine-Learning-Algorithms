```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as pp
```

```python
# a = [-3,-2,-1,0,1,2,3]
a = np.arange(-3,3,0.01)
b = a
a,b = np.meshgrid(a,b)
print(a)
print(b)
```

```python
fig = pp.figure(figsize =(10,7))
# pp.subplot(121)
# axes = fig.gca(projection = '3d')
# print(axes)  #Used to get x and y cordinates of current figure in subplot
# axes.plot_surface(a,b,a**0+5)
# pp.show()
# pp.subplot(122)
axes = fig.gca(projection ='3d')
axes.plot_surface(a,b,a**2+b**2,cmap="rainbow")
pp.show()
```

# to plot surface in 3d use plot_surface function of axes

```python
fig = pp.figure(figsize = (10,7))
axes = fig.gca(projection = '3d')
axes.contour(a,b,a**2+b**2)
pp.show()
```
