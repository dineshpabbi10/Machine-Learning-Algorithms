```python
import matplotlib.pyplot as pp
import numpy as np
```

```python
x = np.array([100,200,50,40,300])
y = np.arange(5)
z = np.array([50,30,20,140,200])
p1 = pp.bar(y,x,0.25,color= "yellow")
# pp.legend("men")
p2 = pp.bar(y+.25,z,0.25,color = "red")
# pp.legend("women")
pp.legend((p1[0], p2[0]), ('Men', 'Women'))
pp.xticks(np.arange(5),('a','b','c','d','e'))
pp.show()
```

# Pie Charts

```python
a = ['a','b','c','d','e']
b = [50,30,20,140,200]
pp.pie(b,labels = a,radius=1,explode = [0,1,0,0,0],shadow = 1)
pp.title("Distribution")
pp.show()
```
